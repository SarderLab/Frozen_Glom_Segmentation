```python
import torch

def load_model_state_dict(filename):
    checkpoint = torch.load(filename,map_location=torch.device('cpu'))
    if "model_dict" in checkpoint:
        return checkpoint["model_dict"]
    else:
        raise ValueError(f"The file {filename} does not contain 'model_dict'.")

def compare_models(model1_dict, model2_dict):
    # Compare layers
    layers1 = set(model1_dict.keys())
    layers2 = set(model2_dict.keys())

    common_layers = layers1.intersection(layers2)
    missing_in_model1 = layers2 - layers1
    missing_in_model2 = layers1 - layers2

    differing_layers = []
    same_layers = []

    # Compare parameters of common layers
    for layer in common_layers:
        if not torch.equal(model1_dict[layer], model2_dict[layer]):
            differing_layers.append(layer)
        else:
            same_layers.append(layer)

    # Print the summary
    print("===== SUMMARY =====")
    print(f"Total layers in model1: {len(layers1)}")
    print(f"Total layers in model2: {len(layers2)}")
    print(f"Common layers: {len(common_layers)}")
    print(f"Layers missing in model1: {len(missing_in_model1)}")
    print(f"Layers missing in model2: {len(missing_in_model2)}")
    print(f"Layers with identical parameters: {len(same_layers)}")
    print(f"Layers with differing parameters: {len(differing_layers)}")
    
    # Detailed differences
    print("\n===== DETAILED DIFFERENCES =====")
    print(f"Layers missing in {model1} : {missing_in_model1} ")
    print(f"\n\nLayers missing in {model2}: {missing_in_model2}")
    print("\n\nLayers with differing parameters:\n", differing_layers)

if __name__ == "__main__":
    model1 = "Laura_glom_train.pth"
    model2 = "non-gs-norm25x.pth"
    model1_dict = load_model_state_dict(model1)
    model2_dict = load_model_state_dict(model2)

    compare_models(model1_dict, model2_dict)
```
